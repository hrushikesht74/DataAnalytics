```r
x <- 1:10
cat(x)
cat('Mean of x is', mean(x))
cat('Median of x is', median(x))
cat('Variance of x is', var(x))
cat('Standard Deviation of x is', sd(x))
```

```r
x <- 1:10
cat(x)
cat('Mean of x is', mean(x),'\n')
cat('Median of x is', median(x),'\n')
cat('Variance of x is', var(x),'\n')
cat('Standard Deviation of x is', sd(x),'\n')
```

```r
print(x)
print(paste('Mean of x is', mean(x)))
print(paste('Median of x is', median(x)))
print(paste('Variance of x is', var(x)))
print(paste('Standard Deviation of x is', sd(x)))
```
